### 2. Analyzing the 'State' Variable

```python
data.state.nunique()
```

```python
data.state.unique()
```

```python
data.state.value_counts()
```

```python
# getting a countplot
plt.figure(figsize =(20,7))

sns.countplot(data.state)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel('States', fontsize=20, color ='Red')
plt.ylabel('Count',fontsize=20,color='Green')
plt.show()
```

```python
#showing churn state-wise
sns.set(style ='whitegrid')
plt.figure(figsize=(20,7))
sns.countplot(data.state, hue = data.churn)

plt.xlabel('States',fontsize = 18, color = 'DarkBlue')
plt.ylabel('Count', fontsize = 18, color = 'DarkOrange');
```

```python
a = data.groupby("state")['churn'].mean()*100
a.sort_values()
```

```python
# line chart of mean churn rate by state
X = data.state.unique()
Y = data.groupby("state")['churn'].mean()

sns.set(style = 'darkgrid')
plt.rcParams['figure.figsize'] = (20,7)
plt.rcParams['lines.linestyle'] = '-'

plt.plot(X,Y, color = 'black', marker = '*', markersize = '10')
plt.title('State Churn Graph', fontsize = 25)
plt.xlabel('States',fontsize = 20)
plt.ylabel('Churn Rate',fontsize = 20);
```

```python
data.groupby('state')['churn'].mean().sort_values(ascending=False).head(5)*100
```

```python
c=pd.crosstab(data.state,data.churn, margins = True)
c['Percentage_Churn'] = c[True]/(c['All'])*100
c.head()
```

```python
type(c)
```

```python
c.info()
```

```python
c.sort_values(by='Percentage_Churn',ascending = False).head()
```
